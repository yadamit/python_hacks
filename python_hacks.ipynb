```python
from IPython.display import clear_output
import time
print("Hello")
for i in range(100):
    print(i)
    time.sleep(.01)
    clear_output(wait=True)
print("Success")
```

```python
import sys
import time
output_stream = sys.stdout

print('Fixed text')
for i in range(10):
    output_stream.write('Update %s\r' % i)
    output_stream.flush()
    time.sleep(.1)
# Optionally add a newline if you want to keep the last update.
output_stream.write('\n')
print('Success')
```

```python
import time
print("Fixed text")
for i in range(100):
    print('Update %d' % i, end='\r')
    time.sleep(.1)
print("Success")
```
